```python
import asyncio
import nest_asyncio
from pathlib import Path

import aiofiles
import aiohttp

nest_asyncio.apply()
```

```python
URL = "http://localhost:8000"
```

## Тестирование ендпоинта `/files/upload`

```python
# Один запрос

async def upload_file(filepath: Path) -> dict[str, str | float]:
    async with aiofiles.open(filepath, "rb") as file:
        content = await file.read()
    form_data = aiohttp.FormData()
    form_data.add_field(
        "uploaded_file",
        content,
        filename=filepath.name,
        content_type="application/octet-stream",
    )
    async with aiohttp.ClientSession(base_url=URL) as session:
        async with session.post(url="/files/upload", data=form_data) as response:
            return await response.json()
```

```python
asyncio.run(upload_file(Path(r"C:\Users\andre\Downloads\22 окт., 11.05_.m4a")))
```

```python
async def upload_file_stream(filepath: Path) -> dict[str, str | float]:
    form_data = aiohttp.FormData()
    form_data.add_field("filename", filepath.name)
    async with aiofiles.open(filepath, "rb") as file:
        content = await file.read()
    form_data.add_field(
        "file",
        content,
        filename=filepath.name,
    )
    async with aiohttp.ClientSession(base_url=URL) as session:
        async with session.post(url="/files/upload-stream", data=form_data) as response:
            return await response.json()
```

```python
asyncio.run(upload_file_stream(Path(r"C:\Users\andre\DIOProjects\dio-audio-summarization\notebooks\file_loading_benchmark\test_files\22 окт., 09.38_.mp3")))
```
